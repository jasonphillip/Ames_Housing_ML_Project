```python
housing = pd.read_csv("data/housing_cleaned.csv")
locations = pd.read_csv("data/locations.csv")
```

```python
# merging Neigborhood_st column
housing = housing.merge(locations[["PID", "Neighborhood_st"]], on="PID", how="left")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f_oneway
from scipy.stats import kruskal

# Load your dataset
# housing = pd.read_csv('your_dataset.csv')

# Step 1: Correlation analysis
# Calculate mean SalePrice, YearBuilt, and GrLivArea for each Neighborhood and Neighborhood_st
neighborhood_means = housing.groupby("Neighborhood").mean()[
    ["SalePrice", "YearBuilt", "GrLivArea"]
]
neighborhood_st_means = housing.groupby("Neighborhood_st").mean()[
    ["SalePrice", "YearBuilt", "GrLivArea"]
]

# Step 2: Kruskal-Wallis H-test
saleprice_neighborhood_groups = [
    group["SalePrice"].values for _, group in housing.groupby("Neighborhood")
]
saleprice_neighborhood_st_groups = [
    group["SalePrice"].values for _, group in housing.groupby("Neighborhood_st")
]

h_stat_saleprice, p_val_saleprice = kruskal(*saleprice_neighborhood_groups)
h_stat_saleprice_st, p_val_saleprice_st = kruskal(*saleprice_neighborhood_st_groups)


# Step 3: Nested model comparison
# Encode categorical variables as dummy variables
housing_with_dummies = pd.get_dummies(
    housing, columns=["Neighborhood", "Neighborhood_st"]
)

# Fit the two linear regression models
model_neighborhood = ols("SalePrice ~ Neighborhood", data=housing_with_dummies).fit()
model_neighborhood_st = ols(
    "SalePrice ~ Neighborhood + Neighborhood_st", data=housing_with_dummies
).fit()

# Perform likelihood ratio test
lr_stat, lr_p_value, lr_df = model_neighborhood.compare_lr_test(model_neighborhood_st)

# Print results

print(
    "Kruskal-Wallis H-test p-values for Neighborhood and Neighborhood_st:",
    p_val_saleprice,
    p_val_saleprice_st,
)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f_oneway
from scipy.stats import kruskal

# Load your dataset
# housing = pd.read_csv('your_dataset.csv')

# Step 1: Correlation analysis
# Calculate mean YearBuilt for each Neighborhood and Neighborhood_st
neighborhood_means = housing.groupby("Neighborhood").mean()["YearBuilt"]
neighborhood_st_means = housing.groupby("Neighborhood_st").mean()["YearBuilt"]

# Step 2: Kruskal-Wallis H-test
yearbuilt_neighborhood_groups = [
    group["YearBuilt"].values for _, group in housing.groupby("Neighborhood")
]
yearbuilt_neighborhood_st_groups = [
    group["YearBuilt"].values for _, group in housing.groupby("Neighborhood_st")
]

h_stat_yearbuilt, p_val_yearbuilt = kruskal(*yearbuilt_neighborhood_groups)
h_stat_yearbuilt_st, p_val_yearbuilt_st = kruskal(*yearbuilt_neighborhood_st_groups)


# Step 3: Nested model comparison
# Encode categorical variables as dummy variables
housing_with_dummies = pd.get_dummies(
    housing, columns=["Neighborhood", "Neighborhood_st"]
)

# Get list of dummy variable column names for each categorical variable
neighborhood_cols = housing_with_dummies.filter(like='Neighborhood_').columns.tolist()
neighborhood_st_cols = housing_with_dummies.filter(like='Neighborhood_st_').columns.tolist()

# Fit the two linear regression models
model_neighborhood = ols("YearBuilt ~ " + ' + '.join(neighborhood_cols), data=housing_with_dummies).fit()
model_neighborhood_st = ols("YearBuilt ~ " + ' + '.join(neighborhood_cols + neighborhood_st_cols), data=housing_with_dummies).fit()

# Perform likelihood ratio test
lr_stat, lr_p_value, lr_df = model_neighborhood.compare_lr_test(model_neighborhood_st)

# Print results
print(
    "Kruskal-Wallis H-test p-values for Neighborhood and Neighborhood_st:",
    p_val_yearbuilt,
    p_val_yearbuilt_st,
)

print("Likelihood ratio test p-value:", lr_p_value)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

# Load your dataset
# housing = pd.read_csv('your_dataset.csv')

# Calculate mean age of the house for each Neighborhood and Neighborhood_st
neighborhood_age_means = housing.groupby("Neighborhood")["YearBuilt"].mean()
neighborhood_st_age_means = housing.groupby("Neighborhood_st")["YearBuilt"].mean()

# Perform two-sample t-test to compare mean age between Neighborhood and Neighborhood_st
t_stat, p_val = ttest_ind(neighborhood_age_means, neighborhood_st_age_means)

# Print results
print("Mean age of the house for each Neighborhood:\n", neighborhood_age_means)
print("Mean age of the house for each Neighborhood_st:\n", neighborhood_st_age_means)
print("Two-sample t-test p-value:", p_val)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway, bartlett

# Load your dataset
# housing = pd.read_csv('your_dataset.csv')

# Calculate variance of YearBuilt for each Neighborhood and Neighborhood_st
neighborhood_var = housing.groupby("Neighborhood")["YearBuilt"].var()
neighborhood_st_var = housing.groupby("Neighborhood_st")["YearBuilt"].var()

# Perform Bartlett test to compare variances
bartlett_stat, p_val = bartlett(neighborhood_var.dropna(), neighborhood_st_var.dropna(), )
print("Bartlett test p-value:", p_val)

# Perform F-test to compare variances
f_stat, p_val = f_oneway(neighborhood_var.dropna(), neighborhood_st_var.dropna())
print("F-test p-value:", p_val)
```
